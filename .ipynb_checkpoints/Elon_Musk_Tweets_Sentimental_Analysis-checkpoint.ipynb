# 1. Import necessary libraries

```python
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.tokenize import TweetTokenizer,word_tokenize
from nltk.stem import PorterStemmer

import warnings
warnings.filterwarnings('ignore')
```

# 2. Load Data

```python
tweets = pd.read_csv('Elon_musk.csv', encoding='mac_roman', error_bad_lines=False)
tweets.head(10)
```

# 3. Data Preprocessing

### Removing empty strings and both leading & trailing characters

```python
tweets_new = [i.strip() for i in tweets.Text]
tweets_new = [i for i in tweets_new if i]
# type(tweets_new)
tweets_new[:11]
```

```python
tweets_df = pd.DataFrame(tweets_new, columns=['text'])
tweets_df
```

```python
corpus = []
```

```python
ps=PorterStemmer()
for i in range(0,1999):
    review = re.sub('@[\w]+','', tweets_df['text'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review] 
    review = ' '.join(review)
    review = review.translate(str.maketrans('','',string.punctuation))
    review = re.sub(r"http\S+", "", review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus_df = pd.DataFrame(corpus, columns=['tweets'])
corpus_df
```

### Joining all into one single string / text

```python
tweets_comb = ''.join(tweets_new)
tweets_comb
```

### Remove twitter usernames handles from text

### Method 1 : using TweetTokenizer from nltk library

```python
tkn = TweetTokenizer(strip_handles=True)
tweets_tkn = tkn.tokenize(tweets_comb)
tweets_tkn_comb = ' '.join(tweets_tkn)
tweets_tkn_comb
```

### Method 2 : using re.sub

```python
tweets_user = re.sub('@[\w]+','',tweets_tkn_comb) #Twitter only allows alphanumeric handles and so [\w] works better for this task
tweets_user
```

### Removing Punctuations

```python
tweets_punc_removed = tweets_user.translate(str.maketrans('','',string.punctuation))
tweets_punc_removed
```

### Removing 'https' / 'url' within text

```python
tweets_removed_url = re.sub(r"http\S+", "", tweets_punc_removed)
tweets_removed_url
```

### Splitting a sentence into a Tokens (words)

```python
tkn_words = word_tokenize(tweets_removed_url, preserve_line=True)
print(tkn_words)
```

### Convert to lower cases

```python
tweet_lower_words = [text.lower() for text in tkn_words]
print(tweet_lower_words)
```

## Converting these words to its root words by stemming

```python
ps = PorterStemmer()
stemmed = [ps.stem(i) for i in tweet_lower_words]
print(stemmed)
```

# 4. Creating a DataFrame of the Tweets for sentiment analysis

```python
sent_df = pd.DataFrame(tweets_new, columns=['sentence'])
sent_df
```

### Getting the negative words affinity values
### Assigning affinity value to be -1 for all the words

```python
afinn_neg = pd.read_csv("negative-words.txt",sep=';', encoding='latin-1',header=None)
afinn_neg
```

```python
afinn_neg[0]
```

```python
afinn_neg[0].isna().sum()
```

```python
afinn_neg.dropna(how='all', axis=1, inplace=True)
afinn_neg
```

```python
afinn_neg.drop([1], axis=1, inplace=True)
afinn_neg.dropna(inplace=True)
afinn_neg
```

## Initially assigning -1 as affinity values for all words

```python
afinn_neg['value']=-1
afinn_neg.columns=['words','value']
afinn_neg.head()
```

```python
afinn_neg=afinn_neg.reset_index()
afinn_neg.drop(['index'], axis=1, inplace=True)
afinn_neg
```

### Getting the positive words affinity values
### Assigning affinity value to be +1 for all the words

```python
afinn_pos = pd.read_csv("positive-words.txt",sep=';', encoding='latin-1',header=None)
afinn_pos
```

```python
afinn_pos=afinn_pos.iloc[:,0]
afinn_pos.dropna(inplace=True)
afinn_pos=afinn_pos.reset_index()
afinn_pos['value']=1
afinn_pos.drop(['index'],axis=1,inplace=True)
afinn_pos.columns=['words','value']
afinn_pos
```

### Concatinating the dataframes to get an affinity dataframe

```python
afinn=pd.concat([afinn_neg,afinn_pos],ignore_index=True)
afinn
```

```python
affinity_scores = afinn.set_index('words')['value'].to_dict()
```

```python
affinity_scores
```

```python
sentiment_lexicon = affinity_scores

def calculate_sentiment(text: str = None):
    sent_score = 0
    if text:
        sentence = nlp(text)
        for word in sentence:
            sent_score += sentiment_lexicon.get(word.lemma_, 0)
    return sent_score
```

## Calculating Sentiment

```python
sent_df['sentiment_value'] = corpus_df['tweets'].apply(calculate_sentiment)
```
